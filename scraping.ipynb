```python
import requests
import time
import random

def fetch_data(url, headers):
    try:
        response = requests.get(url, headers=headers)
        if response.status_code == 429:
            print("Rate limit exceeded. Waiting before retrying...")
            time.sleep(60)  # Wait for 60 seconds before retrying
            return fetch_data(url, headers)  # Retry the request
        else:
            return response
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

standing_url = "https://scholar.google.com/scholar?hl=en&as_sdt=0%2C5&q=Artificial+Intelligence+%28AI%29+and+Machine+Learning+%28ML%29+in+computation&btnG="
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# Randomized delay to avoid hitting rate limits
time.sleep(random.uniform(1, 5))

data = fetch_data(standing_url, headers)
if data:
    print(data.status_code)
else:
    print("Failed to fetch data.")
```
